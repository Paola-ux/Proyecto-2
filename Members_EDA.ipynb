```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from io import StringIO

# Cargar el dataset
df = pd.read_csv("C:/Users/sopor/Desktop/Curso Coding Dojo/Machine Learning/Proyecto 2/Gym Members/gym_members_exercise_tracking.csv")
df["Gender"] = df["Gender"].replace({"Female": 1, "Male": 2})
df["Workout_Type"] = df["Workout_Type"].replace({"Yoga": 1, "HIIT": 2,"Cardio":3, "Strength":4, })
# Funcion para realizar un analisis exploratorio inicial
def eda_inicial(df):
    # Descripcion del conjunto de datos
    # Informacion general sobre el dataset
    print("Descripcion del Conjunto de Datos:")
    print(df.info())  
    # Estadisticas descriptivas basicas
    print("\nEstadisticas Descriptivas:")
    print(df.describe())  

    # Analisis de variables categoricas y valores unicos
    print("\nDistribucion de Variables Categoricas:")
    print(df.select_dtypes(include='object').nunique())  

    # Visualizaciones
    print("\nVisualizaciones:")
    # Histograma para variables numericas
    df.hist(bins=20, figsize=(15, 10))
    plt.suptitle("Histograma de Variables Numéricas")
    plt.show()

    # Grafico de barras para variables categoricas
    for column in df.select_dtypes(include='object'):
        plt.figure(figsize=(10, 6))
        df[column].value_counts().plot(kind='bar')
        plt.title(f"Distribucion de {column}")
        plt.show()

    # Boxplot para detectar outliers
    print("\nIdentificacion de Outliers con Boxplot:")
    for col in df.columns:
        plt.figure(figsize=(6, 4))
        sns.boxplot(x=df[col])
        plt.title(f"Outliers en {col}")
        plt.show()
  

    # Mapa de calor para ver correlaciones
    print("\nMapa de Calor de Correlaciones:")
    plt.figure(figsize=(12, 8))
    sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
    plt.title("Correlacion entre Variables")
    plt.show()

    # Identificacion de valores nulos y mostrar cantidad
    print("\nValores Nulos:")
    print(df.isnull().sum())  
    
    # Identificacion de outliers utilizando Z-score y eliminacion de NaN
    print("\nIdentificacion de Outliers:")
    # Filtrar solo columnas numéricas para Z-score
    numeric_columns = df.select_dtypes(include='number').columns
    
    # Eliminar filas con NaN y calcular el Z-score solo para columnas numericas
    z_scores = stats.zscore(df[numeric_columns].dropna())  
    # Contar los outliers
    outliers = (z_scores > 3).sum(axis=0)  
    print(f"Numero de outliers en cada columna: {outliers}")

    # Funcion para guardar los hallazgos en un archivo
def guardar_hallazgos(hallazgos, filename="hallazgos_eda.txt"):
    with open(filename, 'w') as f:
        f.write(hallazgos)

# Funcion para capturar la informacion del dataset y guardarla en un archivo
def realizar_eda(df):
    # Realizar el analisis exploratorio
    eda_inicial(df)

    # Guardar los hallazgos en un archivo
    hallazgos = "Este es un ejemplo de los hallazgos del EDA.\n"
    
    # Capturar descripcion del dataset y estadisticas descriptivas
    str_io = StringIO()
    
    # Salida de dataset.info()
    df.info(buf=str_io)  
    hallazgos += f"Descripcion del dataset:\n{str_io.getvalue()}\n"

    hallazgos += f"Estadisticas descriptivas:\n{df.describe()}\n"
    
    # Guardar en archivo
    guardar_hallazgos(hallazgos)

# Llamada de la funcion
realizar_eda(df)
```
